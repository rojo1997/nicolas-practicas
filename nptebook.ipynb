```python
# Contar los valores faltantes en cada columna
valores_perdidos = df.isnull().sum()
valores_perdidos
```

```python
#encontrar numero total de valores perdidos  y porcetnaje
total_datos = np.product(df.shape)
total_perdidos = valores_perdidos.sum()
porcentaje_perdidos = (total_perdidos/total_datos) * 100
print(f"Total datos:{total_datos}")
print(f"Total valores perdidos:{total_perdidos}")
print(f"Porcenaje de valores perdidos:{porcentaje_perdidos}")
```

```python
#al haber columnas con info perosnal las eliminamos ya que no son necesarias
columnas_eliminar=['credit_card','phone-number','email','name','company','arrival_date_week_number','reservation_status_date']
df.drop(columnas_eliminar, axis=1, inplace=True)
```

```python
columnas_numericas = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
columnas_numericas
```

```python
columnas_categoricas = df.select_dtypes(include=['object']).columns.tolist()
columnas_categoricas
```

```python
valores_categoricos_perdidos =  [columna for columna in columnas_categoricas if df[columna].isnull().sum()]
for cada_columna in valores_categoricos_perdidos:
    moda_val = df[cada_columna].mode()[0]
    df[cada_columna].fillna(moda_val, inplace=True)
```

```python
valores_numericos_perdidos =  [columna for columna in columnas_numericas if df[columna].isnull().sum()]
for cada_columna in valores_numericos_perdidos:
    media_val = df[cada_columna].mean()
    df[cada_columna].fillna(media_val, inplace=True)
```
